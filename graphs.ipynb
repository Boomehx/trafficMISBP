```python
import pandas as pd
import matplotlib.pyplot as plt

def create_box_plot(csv0, csv1, csv2, csv3, csv4, csv5, column, title, xlabel, ylabel):

    data0 = pd.read_csv(csv0)
    data1 = pd.read_csv(csv1)
    data2 = pd.read_csv(csv2)
    data3 = pd.read_csv(csv3)
    data4 = pd.read_csv(csv4)
    data5 = pd.read_csv(csv5)

    dataToPlt = [
        data0[column], data1[column], data2[column],
        data3[column], data4[column], data5[column]
    ]

    labels = ['Static', 'Estimated BP 75%', 'Estimated BP 50%', 'Estimated BP 25%', 'Simple WBP', 'MIS']

    plt.figure(figsize=(10, 6))
    plt.boxplot(dataToPlt, vert=True, patch_artist=True, notch=True,
                boxprops=dict(facecolor='lightblue'))


    plt.xticks(ticks=range(1, len(labels) + 1), labels=labels, fontsize=11)
    plt.title(title, fontsize=12)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
    
csv0 = r"C:\Users\bobbi\Documents\YEAR 3\sumo3\results\ST05\outputST05.csv"
csv1 = r"C:\Users\bobbi\Documents\YEAR 3\sumo3\results\HIGH75\outputHIGH75.csv"
csv2 = r"C:\Users\bobbi\Documents\YEAR 3\sumo3\results\HIGH50\outputHIGH50.csv"
csv3 = r"C:\Users\bobbi\Documents\YEAR 3\sumo3\results\HIGH25\outputHIGH25.csv"
csv4 = r"C:\Users\bobbi\Documents\YEAR 3\sumo3\results\C6E30LSBP05\outputSBP05.csv"
csv5 = r"C:\Users\bobbi\Documents\YEAR 3\sumo3\results\C6E30LMIS05\outputMIS05.csv"

create_box_plot(csv0, csv1, csv2, csv3, csv4, csv5,
                column="halting",
                title="Comparison of mean travel time across different traffic control algorithms (7200 vehicles/hour)",
                xlabel="Algorithms",
                ylabel="Mean travel time (s) (100-200 runs)")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

csv_file = r"C:\Users\bobbi\Documents\YEAR 3\sumo3\ST05.csv"
output_folder = "perLane"  


df = pd.read_csv(csv_file)
#Ensure output folder exits
columns = df.columns.tolist()
metric_columns = [col for col in columns if '_queueing_time_avg' in col or '_queueing_length_avg' in col]

queueing_time_cols = [col for col in metric_columns if '_queueing_time_avg' in col]
queueing_length_cols = [col for col in metric_columns if '_queueing_length_avg' in col]


def plot_metrics(metrics, title, x_label, y_label, file_name):
    metric_data = df[metrics].copy()
    metric_data = metric_data.melt(var_name="Lane_Metric", value_name="Value")

    metric_data["Lane_Metric"] = (
        metric_data["Lane_Metric"]
        .str.replace('_queueing_time_avg', '', regex=False)
        .str.replace('_queueing_length_avg', '', regex=False)
    )

    plt.figure(figsize=(12, 6))
    sns.boxplot(x="Lane_Metric", y="Value", data=metric_data)
    plt.xticks(rotation=90)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, file_name))
    plt.close()

plot_metrics(queueing_time_cols, 
             title="Mean Queueing Time per Lane for Static Traffic Lights (7200 vehicles/hour)", 
             x_label="Lane ID", 
             y_label="Mean Queueing Time (s)", 
             file_name="QTST.png")

plot_metrics(queueing_length_cols, 
             title="Mean Queueing Length per Lane for Static Traffic Lights (7200 vehicles/hour)", 
             x_label="Lane ID", 
             y_label="Mean Queueing Length (vehicles)", 
             file_name="QLST.png")

print(f"Boxplots saved in '{output_folder}' folder.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp, cumfreq
import os

# Load your CSVs (replace with correct file paths)
data_a = pd.read_csv(r"C:\Users\bobbi\Documents\YEAR 3\sumo3\results\CN25SHORT\outputCN25SHORT.csv") # Algorithm A
data_b = pd.read_csv(r"C:\Users\bobbi\Documents\YEAR 3\sumo3\results\CNSBPSHORT\outputCNSBPSHORT.csv")  # Algorithm B

algorithm_a = data_a["halting"]
algorithm_b = data_b["halting"]

# Perform KS test
statistic, p_value = ks_2samp(algorithm_a, algorithm_b)
print(f"Kolmogorov-Smirnov Test:\nStatistic = {statistic:.4f}, p-value = {p_value:.4g}")

bins = 50
cdf_a = cumfreq(algorithm_a, numbins=bins)
cdf_b = cumfreq(algorithm_b, numbins=bins)

x_a = cdf_a.lowerlimit + np.linspace(0, cdf_a.binsize * cdf_a.cumcount.size, cdf_a.cumcount.size)
x_b = cdf_b.lowerlimit + np.linspace(0, cdf_b.binsize * cdf_b.cumcount.size, cdf_b.cumcount.size)
y_a = cdf_a.cumcount / max(cdf_a.cumcount)
y_b = cdf_b.cumcount / max(cdf_b.cumcount)

plt.figure(figsize=(8, 6))
plt.plot(x_a, y_a, label="Estimated BP (25%)", color="blue")
plt.plot(x_b, y_b, label="Simple WBP", color="orange")
plt.title("Cumulative Distribution Function (CDF) of Halted Vehicles")
plt.xlabel("Number of Halted Vehicles")
plt.ylabel("Cumulative Probability")
plt.legend()
plt.grid(True)
plt.tight_layout()


output_path = r"C:\Users\bobbi\Documents\YEAR 3\sumo3\Labled BoxPlots\halt_cdf_EQ25VSWBP.png"
#plt.savefig(output_path, dpi=300)

# Show the plot
plt.show()
```
