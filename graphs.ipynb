```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration
csv_file = r"C:\Users\bobbi\Documents\YEAR 3\sumo3\AM752.csv"
output_folder = "boxplots"  # Folder to save the plots

# Load the aggregated CSV
df = pd.read_csv(csv_file)

# Create output folder if it doesn't exist
import os
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Extract relevant columns
columns = df.columns.tolist()
metric_columns = [col for col in columns if '_queueing_time_avg' in col or '_queueing_length_avg' in col]

# Separate metrics for plotting
queueing_time_cols = [col for col in metric_columns if '_queueing_time_avg' in col]
queueing_length_cols = [col for col in metric_columns if '_queueing_length_avg' in col]

# Function to plot boxplots for a given set of metrics
def plot_metrics(metrics, title, file_name):
    metric_data = df[metrics]
    metric_data = metric_data.melt(var_name="Lane_Metric", value_name="Value")  # Melt for seaborn compatibility
    
    plt.figure(figsize=(12, 6))
    sns.boxplot(x="Lane_Metric", y="Value", data=metric_data)
    plt.xticks(rotation=90)
    plt.title(title)
    plt.xlabel("Lane and Metric")
    plt.ylabel("Value")
    plt.tight_layout()
    
    plt.savefig(os.path.join(output_folder, file_name))
    plt.close()

# Generate boxplots
plot_metrics(queueing_time_cols, "Boxplots for Queueing Time", "QT752.png")
plot_metrics(queueing_length_cols, "Boxplots for Queueing Length", "QL752.png")

print(f"Boxplots saved in {output_folder}")
```
